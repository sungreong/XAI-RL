```python
import gym
env = gym.make("Breakout-v4")  # Deterministic
```

```python
s = env.reset()
s , _,_,_ = env.step(1)
```

```python
s.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(s)
plt.savefig("temp.png")
plt.show()
```
